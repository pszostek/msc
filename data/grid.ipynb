```python
from matplotlib.pyplot import figure, grid, imshow, show, colorbar
import matplotlib.pyplot as plt
from numpy import array
from pylab import *
from matplotlib import colors
from collections import defaultdict
import sys
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.colors import LogNorm
from matplotlib.ticker import MultipleLocator
import sys
```

```python
def get_maximum_datapoints(list_):
    max_val = 0
    for elem in list_:
        if elem[2] > max_val:
            max_val = elem[2]
    ret = set()
    for elem in list_:
        if elem[2] == max_val:
            ret.add(elem)
    return ret
```

```python
def mark_points(points):
    plt.plot([elem[0] for elem in points], [elem[1] for elem in points], 'bx', markersize=10.0, markeredgewidth=2)
```

```python
def make_matrix(list_):
    import math
    values_dict = defaultdict(dict)
    
    left_extent = min([elem[0] for elem in list_])
    right_extent = max([elem[0] for elem in list_])
    top_extent = max([elem[1] for elem in list_])
    bottom_extent = min([elem[1] for elem in list_])
    
    for elem in list_:
        values_dict[elem[0]][elem[1]] = elem[2]
    values = []
    for d2 in sorted([elem[1] for elem in list_], reverse=True):
        this_vals = []
        for d1 in sorted([elem[0] for elem in list_]):
            this_vals.append(values_dict[d1][d2])
        values.append(this_vals)
    return numpy.matrix(values)
```

```python
def get_data(filename):
    ret = []
    lines = open(filename, 'r').readlines()
    kernel = -1
    for line in lines:
            parts = line.split(' ')
            c=float(parts[0])
            g=float(parts[1])
            rate=float(parts[2])
            ret.append((c,g,rate))
    return ret
            
            
```

```python
def plot(data, minval, maxval, x, y, title_):
    left_extent = min([elem[0] for elem in data])
    right_extent = max([elem[0] for elem in data])
    top_extent = max([elem[1] for elem in data])
    bottom_extent = min([elem[1] for elem in data])

    extent = (left_extent, right_extent, bottom_extent, top_extent)
    data_points = make_matrix(data)
    maximum_data_points = get_maximum_datapoints(data)
    print(maximum_data_points)
  
    figure(figsize=(12,12), dpi=450)
    title(title_, fontsize=23)
    norm = colors.Normalize(vmin=minval, vmax=maxval)
    mark_points(maximum_data_points)
    im = plt.imshow(data_points, norm=norm, extent=extent, interpolation='nearest', cmap="gnuplot2")
    im.get_axes().tick_params(axis='both', which='major', labelsize=15)
    im.get_axes().tick_params(axis='both', which='minor', labelsize=8)
    axes=im.get_axes()
    axes.set_xlabel('C', fontsize=23)
    axes.set_ylabel('gamma', fontsize=23)
    divider = make_axes_locatable(axes)
    cax = divider.append_axes("right", size="5%", pad=0.4)
    bar = plt.colorbar(im, cax=cax)
    for t in axes.get_yticklabels():
        t.set_fontsize(20)
    for t in axes.get_xticklabels():
        t.set_fontsize(20)
    for t in bar.ax.get_yticklabels():
        t.set_fontsize(20)
    
    plt.show()
```
